## Description of the Context: Bounded Knapsack Problem

You are given a set of items, each with an associated value and weight, and a knapsack with a maximum weight capacity. Each item is available a certain number of times. The goal is to determine the most valuable combination of items to place in the knapsack without exceeding its weight capacity.

**Problem Parameters:**

- **Set of Items:** There is a finite set of items, often denoted as $I = \{1, 2, \ldots, n\}$.

- **Item Properties:**
  - Each item $i$ has an associated value $v_i$ (positive integer number), representing the profit gained by including the item in the knapsack.
  - Each item $i$ has an associated weight $w_i$ (positive integer number), indicating the weight of the item.
  - Each item $i$ has an associated maximum quantity $b_i$ (a non-negative integer), specifying the maximum number of times that item can be selected.

- **Knapsack Capacity:** The knapsack has a maximum weight capacity, denoted as $W$ (a positive integer number).

**Objective:**

The objective of the bounded knapsack problem is to determine a selection of items, taking into account the quantity bounds, that maximizes the total value of the selected items while ensuring that the total weight does not exceed the knapsack's capacity.

**Q1.** Write a function `knapdyn(values, weights, quantities, capacity)` which returns the maximum value of a bounded knapsack problem.

```python
#Write your code here
def knapdyn(values, weights, quantities, capacity):
    n = len(values)
    
    # DP table: dp[w] = maximum value achievable with weight w
    dp = [0] * (capacity + 1)
    
    # Process each item type
    for item_idx in range(n):
        value = values[item_idx]
        weight = weights[item_idx]
        quantity = quantities[item_idx]
        
        # Traverse DP table backwards to avoid using same item multiple times
        for w in range(capacity, weight - 1, -1):
            # Try taking k copies of this item (from 1 to quantity)
            for k in range(1, quantity + 1):
                total_weight = k * weight
                total_value = k * value
                
                # Check if we can take k items without exceeding capacity
                if total_weight <= w:
                    dp[w] = max(dp[w], dp[w - total_weight] + total_value)
                else:
                    break
    
    return dp[capacity]
```

```python
#Write your code here
def knapdyn(values, weights, quantities, capacity):
    n = len(values)
    memo = {}
    
    def dp(item_idx, remaining_capacity):
        # Base cases
        if item_idx == n or remaining_capacity == 0:
            return (0, 0)
        
        # Check if already computed
        if (item_idx, remaining_capacity) in memo:
            return memo[(item_idx, remaining_capacity)]
        
        value = values[item_idx]
        weight = weights[item_idx]
        quantity = quantities[item_idx]
        
        # Try taking 0 to quantity items of current type
        max_value = 0
        best_k = 0
        
        for k in range(quantity + 1):
            total_weight = k * weight
            
            # Check if we can take k items
            if total_weight <= remaining_capacity:
                future_value, _ = dp(item_idx + 1, remaining_capacity - total_weight)
                current_value = k * value + future_value
                
                if current_value > max_value:
                    max_value = current_value
                    best_k = k
            else:
                break
        
        memo[(item_idx, remaining_capacity)] = (max_value, best_k)
        return (max_value, best_k)
    
    # Get optimal value and reconstruct solution
    optimal_value, _ = dp(0, capacity)
    
    # Reconstruct solution by tracing through DP decisions
    solution = [0] * n
    current_capacity = capacity
    
    for item_idx in range(n):
        if (item_idx, current_capacity) in memo:
            _, best_k = memo[(item_idx, current_capacity)]
            solution[item_idx] = best_k
            current_capacity -= best_k * weights[item_idx]
    
    return {
        'maximum_value': optimal_value,
        'solution_vector': solution,
        'capacity_used': capacity - current_capacity,
        'capacity_remaining': current_capacity
    }
```

# III. Knapsack by Branch and Bound

Branch and bound is a general algorithmic technique used for solving optimization problems, including the knapsack problem. It systematically explores a search tree of potential solutions, pruning branches that cannot lead to an optimal solution.

The primary steps in the branch and bound approach are as follows:

 - **Branching:** The problem space is divided into smaller subproblems, typically through binary choices. For example, in the knapsack problem, each item can be either included (1) or excluded (0).

 - **Bounding:** At each node in the search tree, an upper bound on the optimal solution is computed. If the bound is worse than the current best solution found, that branch of the tree is pruned, as it cannot lead to an optimal solution.

 - **Backtracking:** The algorithm explores promising branches while keeping track of the best solution found so far. When it reaches a leaf node, it updates the best solution if a better one is found.

**Q2.** Write a function `knapBnB(values, weights, quantities, capacity)`  which returns the maximum value of a bounded knapsack problem with a branch and bound approach.

```python
#Write your code here
def knapBnB(values, weights, quantities, capacity):
    n = len(values)
    best_value = 0
    
    def fractional_bound(item_idx, current_weight, current_value): ##it's like fractional knsap
        if current_weight >= capacity:
            return 0
        
        bound = current_value
        remaining_capacity = capacity - current_weight
        
        # Create list of remaining items with their efficiency ratios
        items = []
        for i in range(item_idx, n):
            if weights[i] > 0:
                ratio = values[i] / weights[i]
                items.append((ratio, values[i], weights[i], quantities[i]))
        
        # Sort by value-to-weight ratio (greedy)
        items.sort(reverse=True, key=lambda x: x[0])
        
        # Calculate fractional bound
        for ratio, value, weight, quantity in items:
            if remaining_capacity <= 0:
                break
            
            max_weight = quantity * weight
            
            if max_weight <= remaining_capacity:
                # Take all available items
                bound += quantity * value
                remaining_capacity -= max_weight
            else:
                # Take fractional amount
                fraction = remaining_capacity / weight
                bound += fraction * value
                remaining_capacity = 0
        
        return bound
    
    def branch_and_bound(item_idx, current_weight, current_value, current_solution):
        nonlocal best_value
        
        # Base case: processed all items
        if item_idx == n:
            if current_value > best_value:
                best_value = current_value
            return
        
        # Calculate upper bound for this node
        upper_bound = fractional_bound(item_idx, current_weight, current_value)
        
        # Pruning: if upper bound is not better than best, skip this branch
        if upper_bound <= best_value:
            return
        
        # Branch: try taking different quantities of current item
        value = values[item_idx]
        weight = weights[item_idx]
        quantity = quantities[item_idx]
        
        # Try taking k items (from 0 to quantity)
        for k in range(quantity + 1):
            total_weight = k * weight
            
            # Check if adding k items fits in capacity
            if current_weight + total_weight <= capacity:
                new_solution = current_solution + [k]
                ##recursive call
                branch_and_bound(
                    item_idx + 1,
                    current_weight + total_weight,
                    current_value + k * value,
                    new_solution
                )
            else:
                # If k items don't fit, larger k won't fit either
                break
    
    # Start branch and bound from root
    branch_and_bound(0, 0, 0, [])
    
    return best_value
```
